```python
import pandas as pd
import geopandas as gpd

afr_gadm_path = '/content/drive/MyDrive/GeoSpatial/DeepTech-Data/afr_gadm/afr_g2014_2013_0.shp'
world_covid_path = '/content/drive/MyDrive/GeoSpatial/DeepTech-Data/world-covid-19/world-covd-19.shp'

afr_gadm_gdf = gpd.read_file(afr_gadm_path)
world_covid_gdf = gpd.read_file(world_covid_path)

print("African GADM GeoDataFrame loaded:")
print(afr_gadm_gdf.head())
print("\nWorld COVID-19 GeoDataFrame loaded:")
print(world_covid_gdf.head())
```

```python
# Reproject the world_covid_gdf to match the CRS of afr_gadm_gdf
world_covid_gdf = world_covid_gdf.to_crs(afr_gadm_gdf.crs)

# Perform the spatial clip operation
africa_covid_gdf = gpd.clip(world_covid_gdf, afr_gadm_gdf)

print("\nAfrican COVID-19 GeoDataFrame (clipped):")
print(africa_covid_gdf.head())
```

```python
import os

# You can save the clipped GeoDataFrame if needed
output_path = '/content/drive/MyDrive/GeoSpatial/DeepTech-Data/africa-covid-19/africa-covid-19.shp'

# Create the directory if it doesn't exist
output_dir = os.path.dirname(output_path)
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Created directory: {output_dir}")

africa_covid_gdf.to_file(output_path)
print(f"\nAfrican COVID-19 GeoDataFrame saved to: {output_path}")
```

```python
# Convert 'date' column to datetime objects
africa_covid_gdf['date'] = pd.to_datetime(africa_covid_gdf['date'])

# Extract month and year
africa_covid_gdf['year'] = africa_covid_gdf['date'].dt.year
africa_covid_gdf['month'] = africa_covid_gdf['date'].dt.month

print("African COVID-19 GeoDataFrame with year and month:")
print(africa_covid_gdf.head())
```

```python
import pandas as pd
import geopandas as gpd # Ensure geopandas is imported

# Filter for the year 2020
africa_covid_2020 = africa_covid_gdf[africa_covid_gdf['year'] == 2020].copy()

# Aggregate cumulative cases by month and country
# Sort by date to ensure the last entry is the most recent cumulative value
africa_covid_2020 = africa_covid_2020.sort_values(by=['key_gadm', 'date'])

# Group by country and month and select the last entry in each group
cumulative_cases_monthly = africa_covid_2020.groupby(['key_gadm', 'month']).tail(1).reset_index()

# Select the relevant columns for the final GeoDataFrame
# Include 'geometry' from the cumulative_cases_monthly as it came from africa_covid_2020 which was a GeoDataFrame
cols_to_keep = [col for col in cumulative_cases_monthly.columns if col != 'geometry'] + ['geometry']
africa_covid_2020_monthly_cumulative = cumulative_cases_monthly[cols_to_keep]


# Convert the DataFrame to a GeoDataFrame, explicitly setting the geometry column
africa_covid_2020_monthly_cumulative_gdf = gpd.GeoDataFrame(africa_covid_2020_monthly_cumulative, geometry='geometry', crs=africa_covid_gdf.crs)


print("\nAfrican COVID-19 Monthly Cumulative Cases for 2020 (with geometry):")
print(africa_covid_2020_monthly_cumulative_gdf.head())
```

```python
# Save the monthly aggregated GeoDataFrame for 2020
output_path_monthly_2020 = '/content/drive/MyDrive/GeoSpatial/DeepTech-Data/africa-covid-19-monthly-2020/africa-covid-19-monthly-2020.shp'

# Create the directory if it doesn't exist
output_dir_monthly_2020 = os.path.dirname(output_path_monthly_2020)
if not os.path.exists(output_dir_monthly_2020):
    os.makedirs(output_dir_monthly_2020)
    print(f"Created directory: {output_dir_monthly_2020}")

africa_covid_2020_monthly_cumulative_gdf.to_file(output_path_monthly_2020)
print(f"\nAfrican COVID-19 Monthly Aggregated GeoDataFrame for 2020 saved to: {output_path_monthly_2020}")
```
